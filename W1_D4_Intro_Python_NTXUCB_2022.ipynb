## Recorrer/iterar una lista usando ciclos *for* y *wile*

- La primera identifica el bucle 
- define un indice, que cambia de valor en cada paso a travez del bucle.
- Despu√©s de la linea de identificaci√≥n viene el grupo de comandos 


`i = 0 -> i = 1 -> i = 2 -> ......... i = 9 `



image.png

```python
# dos ejercicio, string,bool,floats,int con while y for
floats = [0.3, 1.2, 3.4, 2.6, 7.2, 5.6, 4.3]
enteros = list([30, 55, 76, 90, 78, 85])
```

- Repaso de la funcion *len()*

```python
print(len(floats))
```

```python
# iterar con la funcion 'range()'
for indice in range(len(floats)): 
  #print(indice)
  print(floats[indice])

print('fin de ciclo') #mostrar indice
```

```python
#print("indice | valor")
for indice in range(7): 
  print(indice, floats[indice])

print('fin de ciclo') #mostrar indice
```

```python
# iterar con la funcion 'enumerate()'
print("elementos de la lista con enumerate()")
list(enumerate(floats))
```

```python
# ejemplo: mostrar la posicion de los valores menores a 3
# floats = [0.3, 1.2, 3.4, 2.6, 7.2, 5.6, 4.3]
for i,valor in enumerate(floats):
  #print(i, valor)
  if valor < 3:
    print('el numero:', valor,' esta en la posicion ', i)
  
```

###Ciclo While


*   Similares al ciclo for
*   El ciclo **while** continua hasta astisfacer una condicion




image.png

```python
contador = 0
while contador < 10:
  #Codigo dentro del cilo while
  #Codigo dentro del cilo while
  print(contador)
  contador = contador + 1
  
  
  


#Codigo fuera del cilo while
```

```python
c = [32, 6, 45, 76, 23]
```

```python
# graficos explicativos
x = 0
while x < 5:
  print(c[x])
  x = x + 1 # x += 1
```
