# Uso de la libreria Pandas

```python
#importar la libreria pandas
import pandas as pd

# lista de strings
lst = ['PreFrontal', 'Frontal', 'Temporal', 'Occipital', 'Parietal']
  
# Creando un dataframe (tabla)
df = pd.DataFrame(lst) 
df # print(df)
```

## Creando un DataFrame (tabla) usando la funci√≥n *zip()*

```python
# lista de enteros
lst2 = [11, 22, 85, 44, 55]

# Creando un dataframe usando 2 listas
df = pd.DataFrame(list(zip(lst, lst2)),
               columns =['nombre', 'val'])
df
```

### Comparaci√≥n con el uso de un ciclo ***for***

```python
for indice in range(len(lst)): 
  print(lst[indice], lst2[indice])
```

```python
# Creamos un grafico a partir de la tabla
import matplotlib.pyplot as plt

# titulo del gr√°fico
plt.title("Gr√°fico 1", fontdict={'family': 'serif', 
                    'color' : 'blue',
                    'weight': 'bold',
                    'size': 18})


# nombre a los ejes
plt.xlabel("Nombres")
plt.ylabel("Valores")

plt.plot(df['nombre'], df['val']) # eje X, eje y
plt.show()

#df["val"].plot()
```

```python
# lista de nombre carrera y promedio
nombre = ["Manuel", "Richard", "Andres", "Adriana", "Pedro", "Hernan"]
carrera = ["Psicologia", "Matematica", "Ing Biomedica", "Ing. Industrial", "Ing. Sistemas", "Ing. Comercial"]
promedio = [90, 20, 70, 98, 35, 60]
edad = [21,24,43,34, 25, 23]

# Diccionario de listas
dict = {'nombre': nombre, 'carrera': carrera, 'promedio': promedio, 'edad': edad} 

df = pd.DataFrame(dict)

df
#print(df)
```

```python
#df.plot()

plt.title("Gr√°fico 2")

plt.xlabel("Nombre")
plt.ylabel("Promedio")
plt.plot(df['nombre'], df['promedio'], df['edad']) # x , y
plt.legend(['Promedio', 'edad'])
plt.show() 
```
