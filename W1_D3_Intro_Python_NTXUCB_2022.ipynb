### Aprendiendo `enumerate` y `zip`

````python
# Una lista con productos de mercado, en otra lista precio de productos
productos = ['Coca Cola','Nutella','Powerade','Baton']
precios = [10,70,13,2]
````

````python
zip(productos,precios)
````

````python
# Si queremos ver que es lo que zip hace, lo volvemos una lista
list(zip(productos,precios))
````

````python
precios_productos = list(zip(productos,precios))
for producto,precio in precios_productos:
  print(producto,precio)
````

````python
enumerate(productos,precios)
````

````python
precios_productos =list(enumerate(zip(productos,precios)))
for i,(producto,precio) in precios_productos:
  print(i,producto,precio)
````

## Loops

### For Loops

- Los *For Loops* son declaraciones de Python que nos permiten **iterar** sobre distintos tipos de objetos
- Muchos objetos en Python son iterables
  - Por ejemplo, cada elemento de una lista
- Podemos usar los *for loops* para ejecutar bloques de codigo por cada iteraci√≥n que tengamos

*Diagrama de flujo obtenido de Programiz.com

image.png

#### Intro

````python
# Estructura b√°sica
numeros = [1,2,3,4]

for numero in numeros:
  print(numero)
````

````python
# El nombre del iterador puede ser el que tu quieras
numeros = [1,2,3,4]

for elefante in numeros:
  print(elefante)
````

````python
# Usando range()
numeros = list(range(10))

for num in numeros:
  print(num)
````

````python
# Usando range directamente

for num in range(10):
  print(num)
````

````python
# No necesito usar la variable iteradora SI O SI

for i in range(10):
  print('Hola')
````

Un truco cuando tengamos una lista

````python
trastornos_personalidad = ['Ezquizot√≠pico','L√≠mite','Obsesivo Compulsivo','Antisocial','Narcisista']

for i in range(len(trastornos_personalidad)):
  print(i)
  print(trastornos_personalidad[i])
````

#### Realizando bloques de codigo dentro de un *for loop*

````python
# Queremos saber si nuestros amigos pueden viajar al exterior sin permiso legal
edades_amigos = [19,17,23,15,23,21]

for edad in edades_amigos:
  # Viendo si son mayores de edad
  if edad >= 18:
    print('Mayor de edad:',edad)
  else:
    print('Menor de edad:',edad)
````

````python
# Queremos saber nuestra edad total
edades_amigos = [19,17,23,15,23,21]
edad_total = 0

for edad in edades_amigos:
  #Sumando las edades en cada iteracion
  edad_total = edad_total + edad

print('Nuestra edad total:',edad_total)
````

````python
# Ejercicio 3
# Itera esta lista e imprime las variables divididas entre 2
numeros = [424,232,646,834,748,948]

for num in numeros:
  print(num/2)
````

### While Loops

- Los *While Loops* van a continuar ejecutando el bloque de c√≥digo **mientras** una condici√≥n sea verdadera
- Nos sirven mucho cuando solo nos importa la condici√≥n y no el n√∫mero de iteraciones

*Diagrama de flujo obtenido de Programiz.com

image.png

#### Intro

````python
# Estructura b√°sica
variable = 0

while variable < 5:
  print(f'Ahora mismo nuestra variable vale {variable}')
  variable = variable + 1 # Si nos olvidamos de esta linea entraremos en un loop infinito
````

````python
# Un peque√±o truco
variable = 0

while variable < 5:
  print(f'Ahora mismo nuestra variable vale {variable}')
  variable += 1
````

````python
variable = 5

while variable > 0:
  print(f'Ahora mismo nuestra variable vale {variable}')
  variable -= 1
else:
  print('Ahora nuestra variable NO ES mayor que 0')
  #print(variable)
````

#### Realizando *While Loop* con 2 variables

````python
n = int(input('Ingresa el limite: '))
i = 0

while i < n:
  print(f'Iteracion {i}')
  i += 1

else:
  print(f'Fin del Loop, el valor de i es: {i}')
````

#### Pass, continue, break

Estas declaraciones a√±aden funcionalidad a nuestros bloques de c√≥digo



```
pass: no hace nada
continue: Vuelve a correr el c√≥digo dentro de la iteraci√≥n desde arriba
break: Detiene el funcionamiento del Loop
```



````python
numeros = [1,2,3,4]

for num in numeros:
  pass
````

````python
# Imprimiendo solo valores pares
i = 0

while i < 10:
  i += 1

  if (i % 2) != 0:
    continue
    
  else:
    print(i)
````

````python
# Deteniendo el loop bajo alguna condici√≥n

i = 0

while i < 10000000000:
  i += 1

  if i == 7:
    break
    
  else:
    print(i)
````
